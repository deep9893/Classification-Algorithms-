```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
X= pd.get_dummies(df.drop('Exited',axis=1), drop_first=True)
```

```python
y= df['Exited']
```

# train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

X_train= sc.fit_transform(X_train)

X_test=sc.transform(X_test)
```

# modelling

```python
from keras.models import Sequential
```

```python
ann = Sequential()
```

## adding input layers and first hidden layers

```python
from keras.layers import Dense
```

```python
ann.add(Dense(input_dim= 11, units= 6, kernel_initializer= 'uniform',activation = 'relu'))
```

```python
# adding second hidden layer
```

```python
ann.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# adding output layer
ann.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

# compailing ANN

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics= ['accuracy'])
```

# training the Ann on training set

```python
ann.fit(X_train,y_train,epochs =100,batch_size = 32)
```

```python
y_train
```

# prediction and evaluation

```python
train_prediction = ann.predict(X_train)
test_prediction = ann.predict(X_test)
```

```python
test_prediction
```

```python
test_prediction = (test_prediction>=0.5)
train_prediction = (train_prediction>=0.5)
```

```python
test_prediction
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

print('test accuracy:', accuracy_score(y_test,test_prediction))
print('train accuracy:', accuracy_score(y_train,train_prediction))
confusion_matrix(y_test,test_prediction)
```

# cross validation 

```python
def built_cross_classifier():
    classifier = Sequential()
    classifier.add(Dense(input_dim=11, units=6,kernel_initializer= 'uniform', activation = 'relu'))
    classifier.add(Dense(units=6,kernel_initializer= 'uniform', activation = 'relu'))
    classifier.add(Dense(units=1,kernel_initializer= 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics= ['accuracy'])   
    return classifier
```

```python
from scikeras.wrappers import KerasClassifier
```

```python
classifier = KerasClassifier(built_cross_classifier, batch_size = 10,epochs = 50)
```

```python
from sklearn.model_selection import cross_val_score
score= cross_val_score(classifier,X,y,cv = 5)
print (score)
score.mean()
```

```python
classifier = KerasClassifier(built_cross_classifier())
```

```python
estimator = classifier
param_grid = {'batch_size':[10,32], 'epochs':[50,100],'optimizer':['adam','rmsprop']}
```

```python
from sklearn.model_selection import GridSearchCV
hp = GridSearchCV(estimator , param_grid, scoring= 'accuracy', cv=5)

hpt= hp.fit(X_train,y_train)
```
